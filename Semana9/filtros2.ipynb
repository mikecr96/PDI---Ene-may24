```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI - Ene-may24/images/lena.jpg", 0)
```

```python
plt.imshow(img, "gray")
plt.show()
```

```python
def generate_kernel(size):
    return 1/size**2 * np.ones((size, size))
```

```python
lena_filtropromedio = cv.filter2D(img, -1, generate_kernel(5))
```

```python
plt.imshow(lena_filtropromedio, "gray")
plt.show()
```

### Desenfoque Gaussiano

```python
lena_filtro_gauss = cv.GaussianBlur(img, (5,5), sigmaX=1)
```

```python
plt.imshow(lena_filtro_gauss, "gray")
plt.show()
```

```python
plt.imshow(np.hstack((lena_filtropromedio, lena_filtro_gauss)), "gray")
```
