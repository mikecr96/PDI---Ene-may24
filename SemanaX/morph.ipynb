### Erosi√≥n

```python
jota = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI - Ene-may24/images/j.png", 0)
```

```python
plt.imshow(jota, "gray")
plt.show()
```

```python
kernel1 = np.ones((3,3), dtype=np.uint8)
kernel2 = np.ones((5,5), dtype=np.uint8)
```

```python
np.unique(jota)
```

```python
jota_erosionada = cv.erode(jota, kernel1, iterations=2)
plt.imshow(np.hstack((jota, jota_erosionada)), "gray")
plt.title("Img. erosionada con kernel chico 2 iteraciones")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
jota_erosionada = cv.erode(jota, kernel2, iterations=1)
plt.imshow(np.hstack((jota, jota_erosionada)), "gray")
plt.title("Img. erosionada con kernel grande 1 iteraci√≥n")
plt.xticks([])
plt.yticks([])
plt.show()
```

---
### Dilataci√≥n

```python
jota_diltada = cv.dilate(jota, kernel1, iterations=2)
plt.imshow(np.hstack((jota, jota_diltada)), "gray")
plt.title("Img. erosionada con kernel chico 2 iteraciones")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
jota_diltada2 = cv.dilate(jota, kernel2, iterations=1)
plt.imshow(np.hstack((jota, jota_diltada2)), "gray")
plt.title("Img. dilatada con kernel grande 1 iteraci√≥n")
plt.xticks([])
plt.yticks([])
plt.show()
```
