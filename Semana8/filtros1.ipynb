```python
def generate_kernel(size):
    return 1/size**2 * np.ones((size, size))
```

```python
generate_kernel(7)
```

```python
img_filtrada1 = cv.filter2D(img, -1, generate_kernel(5))
plt.imshow(img_filtrada1, 'gray')
plt.show()
```

```python
plt.imshow(np.hstack((img, img_filtrada1)), 'gray')
plt.show()
```
