```python
img = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI - Ene-may24/images/sudoku.png", 0)
```

```python
angle = 0.
scale = 10

M = cv.getRotationMatrix2D((img.shape[0] // 2, img.shape[1] // 2), angle, scale)

plt.imshow(cv.warpAffine(img, M, img.shape, borderValue=255), "gray")
plt.show()
```
