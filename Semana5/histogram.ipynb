#### Img. a color

```python
img_color = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI - Ene-may24/images/baboon.jpg")
```

```python
img_color.shape
```

```python
histogram_b = cv.calcHist([img_color], [0], None, [256], (0, 256))
histogram_g = cv.calcHist([img_color], [1], None, [256], (0, 256))
histogram_r = cv.calcHist([img_color], [2], None, [256], (0, 256))
```

```python
# Histograma azul
plt.plot(histogram_b, "b")
plt.title("Histograma azul")
```

```python
# Histograma verde
plt.plot(histogram_g, "g")
plt.title("Histograma verde")
```

```python
# Histograma rojo
plt.plot(histogram_r, "r")
plt.title("Histograma rojo")
```

```python
colors = ("b", "g", "r")

for n, i in enumerate(colors):
    plt.plot(cv.calcHist([img_color], [n], None, [256], (0, 256)), i)
```
