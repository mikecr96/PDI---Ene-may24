#### Ecualizaci√≥n del histograma

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_gray = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI - Ene-may24/images/baboon.jpg", 0)
```

```python
plt.imshow(img_gray, "gray")
plt.show()
```

```python
hist_orig = cv.calcHist([img_gray], [0], None, [256], (0, 256))
```

```python
plt.plot(hist_orig)
plt.title("Histograma de la img orig.")
plt.show()
```

```python
img_eq = cv.equalizeHist(img_gray)
```

```python
plt.imshow(np.hstack((img_gray, img_eq)), "gray")
plt.show()
```

#### Eq. en img. a color

```python
img_color = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI - Ene-may24/images/baboon.jpg")[:, :, ::-1]
```

```python
plt.imshow(img_color)
plt.show()
```

```python
# Ecualizar todos los canales

r, g, b = cv.split(img_color)
r_eq = cv.equalizeHist(r)
g_eq = cv.equalizeHist(g)
b_eq = cv.equalizeHist(b)
img_color_eq = cv.merge((r_eq, g_eq, b_eq))
```

```python
plt.imshow(np.hstack((img_color, img_color_eq)))
plt.show()
```

```python
cv.equalizeHist(img_color)
```

```python
# YCbCr
# Origen RGB destino es YCbCr
img_color_ycrcb = cv.cvtColor(img_color, cv.COLOR_RGB2YCrCb)
```
